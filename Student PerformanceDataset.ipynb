```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/Student Performance dataset.csv")

# Display first 5 records
print("First 5 rows:")
display(df.head())

# Display last 5 records
print("Last 5 rows:")
display(df.tail())
```

```python
# Numerical features (int, float)
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

# Categorical features (object)
categorical_features = df.select_dtypes(include=["object"]).columns.tolist()

# Binary features (columns with only 2 unique values)
binary_features = [col for col in df.columns if df[col].nunique() == 2]

# Ordinal features (manually identified based on meaning)
# Example: education level, grades, study time etc.
ordinal_features = [
    col for col in categorical_features
    if col.lower() in ["parental level of education", "grade", "education"]
]

print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
print("Binary Features:", binary_features)
print("Ordinal Features (manual):", ordinal_features)
```

```python
# Dataset structure and null values
df.info()

# Statistical summary for numerical columns
df.describe()
```

```python
for col in  categorical_features:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
# Target variable (example: final score or performance)
target_column = "math_score"   # change if your dataset has a different target

# Input features
X = df.drop(columns=[target_column])
y = df[target_column]

print("Target Variable:", target_column)
print("Input Features:", X.columns.tolist())
```

```python
rows, columns = df.shape

print("Number of rows:", rows)
print("Number of columns:", columns)

if rows >= 500:
    print("Dataset size is suitable for basic machine learning models.")
else:
    print("Dataset may be small; simple ML models are recommended.")
```

```python
# Missing values count
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)

# Check class imbalance (if classification)
if y.nunique() <= 10:
    print("\nTarget variable distribution:")
    print(y.value_counts(normalize=True))

# Summary observations
print("\nObservations:")
if missing_values.sum() == 0:
    print("- No missing values found.")
else:
    print("- Missing values present; imputation required.")

print("- Check categorical imbalance before model training.")
print("- Feature encoding needed for categorical variables.")
```
